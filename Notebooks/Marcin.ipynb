```python
import json
import regex as re


with open("../dict_wB.json", 'r') as fp:
    data = json.load(fp)

# Lista de nombres de archivos
print(data.keys())
keys = data.keys()
arr = []

# para leer el contenido de uno de los archivos a secas usar data[nombrearchivo]
#Grupo 1
for filename in data.keys():
    intarr = []
    matches = re.findall('ES-\d{3}', \
                data[filename])
   #facturas real y pura
    matches1 = re.findall('(?<=FACTURA SIMPLIFICADA ).*', \
                data[filename])
#ticket
    matches2 = re.findall('(?<=Número de factura).*', \
                data[filename])
    #prueba 1_1 y 1_2
    matches3 = re.findall('FRO\d{11}', \
                data[filename])
    #prueba 5 y prueba 6
    matches4 = re.findall('(?<=NÚMERO\n).*', \
                data[filename])
    #prueba 9 prueba 10 y prueba 11
    matches5 = re.findall('(?<=FACTURA\n).*\d{4} \d{2}', \
                data[filename])
    #ivalineas 2
    #matches6 = re.findall('(?<=\d{8}[A-Z]\n).*\n.*$', \
    #            data[filename])
    #prueba1_3
    matches6 = re.findall('(?<=FACTURA : ).*\d{4}', \
                data[filename])
    #prueba 
    matches7 = re.findall('(?<=\d{8}[A-Z]\n).*\d{3}\/\n\d{1} \d{3}', \
                data[filename])
    
    matches7 = [line.replace("\n","") for line in matches7]
    #prueba 1_4
    matches8 = re.findall('(?<=\d{2}\/\d{2}\/\d{4}\n).*\d{3}\/\d{4}', \
                data[filename])
    #ivanormal3 
    matches9 = re.findall('(?<=FACTURA: ).*[A-Z]\/\d{6}', \
                data[filename])
    #ivalineas3
    matches10 = re.findall('(?<=\d{2}\/\d{2}\/\d{2}\n).*\d{2}\/\d{3}', \
                data[filename])
    #12ocr
    matches11 = re.findall('(?<=Factura: ).*\d', \
                data[filename])
    #ivalineas1?
    matches12 = re.findall('(?<=Numero: ).*\d{3}', \
                data[filename])
    #codigoqr
    
    #para los numeros de teléfono
    
    matches13 = re.findall('9\d \d{3} \d{2} \d{2}\s| 9\d{2} \d{6}\s| 9\d{2} \d{3} \d{3}\s| 9\d{8}\s| 9\d{2} \d{2} \d{2} \d{2}\s', \
                data[filename])
    matches13 = [line.strip().strip('\n') for line in matches13]
    
    #correos electrónicos
    matches14 = re.findall('(?:\w+\.)?\w+@\w+.\w+', \
                data[filename])
    

    
    
    
    #prueba 1_4 antiguo: (?<=\d{8}[A-Z]\n).*\d{3}\/\n\d{1} \d{3}
    
    if matches:
        intarr.append(matches)
    
    if matches1:
        intarr.append(matches1)
    
    if matches2:
        intarr.append(matches2)
    
    if matches3:
        intarr.append(matches3)
    
    if matches4:
        intarr.append(matches4)
        
    if matches5:
        intarr.append(matches5)
    
    if matches6:
        intarr.append(matches6)
        
    if matches7:
        intarr.append(matches7)
    
    if matches8:
        intarr.append(matches8)
        
    if matches9:
        intarr.append(matches9)
        
    if matches10:
        intarr.append(matches10)
        
    if matches11:
        intarr.append(matches11)
        
    if matches12:
        intarr.append(matches12)
        
    if matches13:
        intarr.append(matches13)
        
    if matches14:
        intarr.append(matches14)
        
    arr.append(intarr)
newdic = dict(zip(keys,arr))

#------------------------------------------------------------------------------------------------------------



print(newdic)
with open('num_facturas_marcin.txt', 'w') as f:
    for filename in newdic.keys():
        f.write(f'{filename}: {newdic[filename]}\n')

#for a in newdic.keys():
 #   newdic[a]
```
